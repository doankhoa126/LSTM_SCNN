```python
import pandas as pd  
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from datetime import datetime
from selenium import webdriver
import pandas as pd
import pyodbc
import concurrent.futures
from sqlalchemy import create_engine
from concurrent.futures import ThreadPoolExecutor
```

```python
def exchangeUPCOM():
    driver = webdriver.Chrome()

    url = 'https://www.stockbiz.vn/MarketMovers.aspx?Exchange=UPCOM'
    driver.get(url)

    

    data = [] 

    for loop_count in range(1):
        table = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//*[@id="market_statictiss_content"]/table/tbody/tr/td[1]/table/tbody/tr/td[1]/table')))

        rows = table.find_elements(By.TAG_NAME, 'tr')

        for row in rows:
            cells = row.find_elements(By.TAG_NAME, 'td')
            row_data = [cell.text for cell in cells]
            data.append(row_data)


    df = pd.DataFrame(data)
    df = df.dropna(how='any')
    
    df = df.transpose()

    # Đặt hàng đầu làm tiêu đề cột
    df.columns = df.iloc[0]

    # Bỏ hàng đầu
    df = df[1:]

    # Đặt lại chỉ mục (index) nếu cần
    df.reset_index(drop=True, inplace=True)
    df.columns = [None] * len(df.columns)
    df.columns = ['totalTradedShares','sharesNoTrading','sharesIncreased','sharesDecreased','sharesUnchanged']
    
    df[['sharesInc', 'percentage']] = df['sharesIncreased'].str.split(' / ', expand=True)
    df['perSharesInc'] = df['percentage'].str.rstrip('%')

    df[['sharesDec', 'percentage']] = df['sharesDecreased'].str.split(' / ', expand=True)
    df['perSharesDec'] = df['percentage'].str.rstrip('%')

    df[['sharesUnc', 'percentage']] = df['sharesUnchanged'].str.split(' / ', expand=True)
    df['persharesUnc'] = df['percentage'].str.rstrip('%')
    
    columns_to_drop = ['sharesIncreased', 'sharesDecreased', 'sharesUnchanged','percentage' ]
    df = df.drop(columns=columns_to_drop)
   
    current_date = datetime.now().date()
    df['date'] = current_date
    
    dfUPCOM = df.assign(exchange='UPCOM')
    
    
    print(dfUPCOM)
    
    driver.quit()
    

    return dfUPCOM
```

```python
import pyodbc
import pandas as pd

def insertSQL(df):
    conn = pyodbc.connect('Driver={SQL Server};'
                          'Server=MSI\\SQLEXPRESS;'
                          'Database=dataFinance;'
                          'Trusted_Connection=yes;')
    cursor = conn.cursor()
    
    df['date'] = pd.to_datetime(df['date']).dt.strftime('%Y-%m-%d')


    # Chuyển đổi kiểu dữ liệu cho các cột
    df['perSharesInc'] = df['perSharesInc'].str.replace(',', '.').astype(float)
    df['perSharesDec'] = df['perSharesDec'].str.replace(',', '.').astype(float)
    df['persharesUnc'] = df['persharesUnc'].str.replace(',', '.').astype(float)

    for index, row in df.iterrows():
        sql = f"INSERT INTO ratioExchange (exchange, date, totalTradedShares, sharesNoTrading, sharesInc, perSharesInc, sharesDec, perSharesDec, sharesUnc, persharesUnc) VALUES ('{row['exchange']}', '{row['date']}', {int(row['totalTradedShares'])}, {int(row['sharesNoTrading'])}, {int(row['sharesInc'])}, {row['perSharesInc']}, {int(row['sharesDec'])}, {row['perSharesDec']}, {int(row['sharesUnc'])}, {row['persharesUnc']})"
        cursor.execute(sql)

    conn.commit()
    conn.close()
```

```python
def main():
    with ThreadPoolExecutor(max_workers=3) as executor:
        result_hnx = executor.submit(exchangeHNX)
        result_hose = executor.submit(exchangeHOSE)
        result_upcom = executor.submit(exchangeUPCOM)

    # Lấy kết quả từ hai hàm
    dfHNX = result_hnx.result()
    dfHOSE = result_hose.result()
    dfUPCOM=result_upcom.result()
    # Gộp hai DataFrame thành một DataFrame duy nhất
    df = pd.concat([dfHNX, dfHOSE,dfUPCOM], ignore_index=True)
    insertSQL(df)
    print("insert succes")
   

```

```python
if __name__ == "__main__":
    main()
```
