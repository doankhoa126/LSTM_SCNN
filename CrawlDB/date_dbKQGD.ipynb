```python
def insert_Trading(df):
    conn = pyodbc.connect('Driver={SQL Server};'
                          'Server=MSI\\SQLEXPRESS;'
                          'Database=dataFinance;'
                          'Trusted_Connection=yes;')


    cursor = conn.cursor()

    df = df.replace('-', 0)

    df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')





    def custom_float_conversion(value):
        if isinstance(value, str):
            try:
                return float(value.replace(',', ''))
            except ValueError:
                return value
        elif isinstance(value, int):
            return float(value)
        else:
            return value


    columns_to_convert_to_float = ['openPrice', 'closePrice', 'highPrice', 'lowPrice', 'average', 'changePlusMinus', 'changePercent', 'orderMatchingVol', 'orderMatchingVal', 'putThroughVol', 'putThroughVal', 'totalTransactionVol', 'totalTransactionVal', 'marketCap']

    for column in columns_to_convert_to_float:
        df[column] = df[column].apply(custom_float_conversion)


    for index, row in df.iterrows():
    # Chuyển đổi các giá trị rỗng (dấu '-') thành 0
        for col in df.columns:
            if pd.isna(row[col]) or row[col] == '-':
                row[col] = 0

    for index, row in df.iterrows():
        cursor.execute("INSERT INTO tradingStock (date, ticker,exchange,reference,openPrice,closePrice,highPrice,lowPrice,average,changePlusMinus,changePercent,orderMatchingVol,orderMatchingVal,putThroughVol,putThroughVal,totalTransactionVol,totalTransactionVal,marketCap) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)",
                       row['date'], row['ticker'], row['exchange'], row['reference'], row['openPrice'], row['closePrice'], row['highPrice'], row['lowPrice'], row['average'], row['changePlusMinus'], row['changePercent'], row['orderMatchingVol'], row['orderMatchingVal'], row['putThroughVol'], row['putThroughVal'], row['totalTransactionVol'], row['totalTransactionVal'], row['marketCap'])

    print("Succes")
    

    conn.commit()
    conn.close()
```
